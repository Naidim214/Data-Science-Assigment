```python
#Who has a higher chance of getting Lung Cancer
Gender_Lung_Cancer_counts = df_clean.groupby('GENDER')['LUNG_CANCER'].value_counts().unstack()
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].pie(Gender_Lung_Cancer_counts.loc[1], labels=['Negative', 'Positive'], autopct='%1.1f%%', startangle=90)
ax[0].set_title('Lung Cancer Distribution Among Males')
ax[1].pie(Gender_Lung_Cancer_counts.loc[0], labels=['Negative', 'Postive'], autopct='%1.1f%%', startangle=90)
ax[1].set_title('Lung Cancer Distribution Among Females')

plt.show()
```

```python
#Does smoking cause Lung Cancer?
Smokers_count = df_clean.groupby('SMOKING')['LUNG_CANCER'].value_counts().unstack()
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].pie(Smokers_count.loc[1], labels=['Negative', 'Positive'], autopct='%1.1f%%', startangle=90)
ax[0].set_title('Lung Cancer Distribution Among Smokers')
ax[1].pie(Gender_Lung_Cancer_counts.loc[0], labels=['Negative', 'Postive'], autopct='%1.1f%%', startangle=90)
ax[1].set_title('Lung Cancer Distribution Among Non-Smokers')

plt.show()
```
